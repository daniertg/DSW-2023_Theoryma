```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset
dataset = pd.read_excel('/content/drive/MyDrive/Lomba/DSW 2023/Telco_customer_churn_adapted_v2.xlsx')

# Fitur yang akan diplot terhadap label
features = ['Tenure Months', 'Device Class', 'Games Product', 'Music Product', 'Education Product',
            'Call Center', 'Video Product', 'Use MyApp', 'Payment Method', 'Monthly Purchase (Thou. IDR)',
            'CLTV (Predicted Thou. IDR)']

# Memvisualisasikan persebaran data menggunakan subplot
fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(30, 18))
fig.subplots_adjust(hspace=0.25, wspace=0.7)

for i, feature in enumerate(features):
    row, col = i // 4, i % 4
    sns.countplot(x=feature, hue='Churn Label', data=dataset, ax=axes[row, col])
    axes[row, col].set_title(f'{feature} vs Churn Label')

plt.show()
```
