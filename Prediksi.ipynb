```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical


data = pd.read_excel('/content/drive/MyDrive/Lomba/DSW 2023/Telco_customer_churn_adapted_v2.xlsx')
# Hapus kolom yang tidak diinginkan
columns_to_drop = ["Location", "Longitude", "Latitude", "Customer ID"]
data = data.drop(columns=columns_to_drop, axis=1)

# Mengubah data kategorikal menjadi numerik menggunakan LabelEncoder
le = LabelEncoder()
data['Device Class'] = le.fit_transform(data['Device Class'])
data['Games Product'] = le.fit_transform(data['Games Product'])
data['Music Product'] = le.fit_transform(data['Music Product'])
data['Education Product'] = le.fit_transform(data['Education Product'])
data['Call Center'] = le.fit_transform(data['Call Center'])
data['Video Product'] = le.fit_transform(data['Video Product'])
data['Use MyApp'] = le.fit_transform(data['Use MyApp'])
data['Payment Method'] = le.fit_transform(data['Payment Method'])
data['Churn Label'] = le.fit_transform(data['Churn Label'])

# Tentukan batas rentang Tenure Months
tenure_ranges = [0, 12, 24, 36, 48, 60, 72]

for i in range(len(tenure_ranges) - 1):
    # Filter data berdasarkan rentang Tenure Months
    tenure_min = tenure_ranges[i]
    tenure_max = tenure_ranges[i + 1]
    filtered_data = data[(data['Tenure Months'] >= tenure_min) & (data['Tenure Months'] < tenure_max)]

    # Konversi label Churn menjadi 1 (Yes) atau 0 (No)
    le_churn = LabelEncoder()
    filtered_data['Churn Label'] = le_churn.fit_transform(filtered_data['Churn Label'])

    # Ambil fitur dan label
    X = filtered_data.drop('Churn Label', axis=1).values
    y = filtered_data['Churn Label'].values

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

    # Standardisasi fitur menggunakan StandardScaler
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Konversi label menjadi one-hot encoding
    y_train_encoded = to_categorical(y_train)
    y_test_encoded = to_categorical(y_test)

    # Membangun model ANN
    model = Sequential()
    model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(2, activation='softmax'))

    # Mengompilasi model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Melatih model
    model.fit(X_train, y_train_encoded, epochs=50, batch_size=32, validation_data=(X_test, y_test_encoded))

    # Memprediksi Churn Label menggunakan data uji
    predictions = np.argmax(model.predict(X_test), axis=1)

    # Menghitung akurasi prediksi
    accuracy = accuracy_score(y_test, predictions)
    print(f'Accuracy for Tenure Months in range {tenure_min} to {tenure_max} with Churn Label = Yes: {accuracy}')
```
